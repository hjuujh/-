```
from bs4 import BeautifulSoup 
import requests 

from selenium.webdriver.common.keys import Keys 
import time 
import re 
import pandas as pd
```

```
def getPageLinksWantRange(startPageNo, lastPageNo): 
  links = []
  
  for pageNo in range(startPageNo-1, lastPageNo): 
    url = "http://www.yes24.com/24/category/bestseller?CategoryNumber=001&sumgb=03&PageNumber=" + str(pageNo + 1) 
    req = requests.get(url)
    soup = BeautifulSoup(req.content, 'lxml') 
    # print(soup)
    # booklinks = soup.select('div#yesWrap div#wrapperContent div#bestList table#category_layout tbody tr td.goodsTxtInfo p a[href]') 
    booklinks = soup.select('div#yesWrap div#wrapperContent div#bestList table#category_layout tr td.goodsTxtInfo p.review a[href]') 
    # print(booklinks)

    for booklink in booklinks: 
      link = str(booklink.get('href')) 
      links.append("http://www.yes24.com"+link)
  return links
  # print(links)
    # print(len(links))
```

```
links = getPageLinksWantRange(1, 1)
# print(links)
```
